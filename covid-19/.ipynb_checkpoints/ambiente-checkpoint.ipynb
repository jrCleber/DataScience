```python
class DSAmbiente:
    def __init__(self):
        self._diretorios = '001 - Dados Originais', '002 - Dados Preparados', '003 - Analises', '004 - Dados Upload', '005 - Insigths', '006 - Final', '_versionamento_'
            
    def getDir(self):
        return self._diretorios
    
    def cosntrucao(self, dataset, desktop):
        diretorio = 'Analises'
        home = path.home()
        ambiente_criacao = home / 'Desktop' / 'Python' / diretorio
        print(desktop)
        
        if path.is_dir(ambiente_criacao) == False:
            print('Path inesistente')
            print('Criando Path...')
            try:
                path.mkdir(ambiente_criacao)
            except FileNotFoundError as e:
                print('[WinError 3] O sistema não consegue encontrar o caminho especificado em:')
                print(ambiente_criacao)
        else:       
            chdir(ambiente_criacao)
            print(path.cwd())

            for pasta in self.getDir():
                if not path(ambiente_criacao / pasta).exists():
                    path.mkdir(ambiente_criacao / pasta)
                if pasta == self.getDir()[0]:
                    local = ambiente_criacao / self.getDir()[0]
                    chdir(local)
                    try:
                        cp(dataset, local)
                    except FileNotFoundError as e:
                        print(type(e))
                        print('[WinError 3] O sistema não consegue encontrar o caminho especificado em:\n', dataset)
                        print('Não foi possível copiar o arquivo para:\n', local)
            chdir(desktop)
        
    def dataset_file(self, caminho, desktop): 
        print('SELECIONE O DATASET:\n'+'_'*60)    
        chdir(caminho)
        print(path.cwd())
        list_file = listdir()
        dict_file = {}
        for i in range(len(list_file)):
            dict_file[i] = list_file[i]
        for k,v in dict_file.items():
            print('(%i) para para o Dataset: %s'%(k,v))
        print('_'*60)
        chdir(desktop)
        while True:
            try:
                print('-'*60)
                return caminho / list_file[int(input('Escolha: '))]
            except ValueError as e:
                print(type(e))
                print('Valor inválido')
            except IndexError as e:
                print(type(e))
                print('Índice inválido')
```
