# Trabalhando com o arquivo covid_work_int_2020-06-15.xlsx

```python
covid_work = pd.read_excel(ab.ambiente.file_data_ambiente())
covid_work.head()
```

```python
covid_work.dtypes
```

## Configurando as datas

```python
codid_work_date = covid_work.loc[:, ['data', 'semanaEpi', 'casosNovos','obitosNovos']]
```

```python
codid_work_date['dia'] = codid_work_date.data.dt.day
codid_work_date['mes'] = codid_work_date.data.dt.month
codid_work_date['ano'] = codid_work_date.data.dt.year
```

```python
codid_work_date.head()
```

## Agrupando por data

```python
covid_date_group = codid_work_date.groupby(['ano','mes','semanaEpi','dia']).sum()
covid_date_group.head(20)
```

```python
#exportando para insigths
insigths = pth.home() / 'Desktop' / 'Python' / 'Analises' / '005 - Insigths'
name = 'covid_data_group.xlsx'
if pth(insigths).exists():
    covid_date_group.to_excel(insigths/name, encoding='latin-1', index=True)
```

## Agrupando por regiões

```python
covid_regiao = covid_work.loc[:,['regiao','estado','nomeRegiaoSaude','municipio','casosNovos','obitosNovos','populacaoTCU2019_int']]
covid_regiao_group = covid_regiao.groupby(['regiao','estado','nomeRegiaoSaude','municipio']).agg(['sum','mean','min','max'])
covid_regiao_group.head(10)
```

```python
#exportando para insigths
name = 'covid_regiao_group.xlsx'
if pth(insigths).exists():
    covid_regiao_group.to_excel(insigths/name, encoding='latin-1', index=True)
```
